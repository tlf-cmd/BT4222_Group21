```python
import pandas as pd
```

```python
def summary_viewer(ticker):
    generated_summary = pd.read_csv(f"Generated Summaries/final/{ticker}_top30_summaries.csv",index_col=0)
    generated_summary = generated_summary.rename(columns={'0': 'Generated Summary', '1': 'Reference Summary'})
    criteria_topic_scores = pd.read_csv(f"Scrapers/scraped_data/{ticker}_criteria_topic_scores.csv",index_col=0)
    criteria_topic_scores = criteria_topic_scores.rename(columns={ticker: 'Criteria Topic Score'})

    output = pd.merge(generated_summary, criteria_topic_scores,left_index=True,right_index=True)

    # print the output
    return output
```

```python
summary_viewer('AAPL')
```

```python
summary_viewer('ABBV')
```
